```python
from transformers import AutoTokenizer,AutoModelForMaskedLM
```

```python
tokenizer=AutoTokenizer.from_pretrained(model_name)
```

```python
def tokenization_fun(data):
  return tokenizer(data['text'],padding='max_length',truncation=True)
```

```python
tokenized_data=raw_dataset.map(tokenization_fun)
```

```python
train_dataset_trainer=tokenized_data['train']
test_dataset_trainer=tokenized_data['test']
```

```python
from transformers import AutoModelForSequenceClassification
```

```python
model=AutoModelForSequenceClassification.from_pretrained(model_name,num_labels=2)
```

```python
from transformers import TrainingArguments,Trainer
```

```python
train_args=TrainingArguments(
    output_dir='./model_result',
    num_train_epochs=5,
    #learning_rate=2e-5,
    eval_strategy="epoch",
    save_strategy="epoch",
    per_device_train_batch_size=16,
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",

)
```

```python
def compute(pred):
  labels=pred.label_ids #[0,1]
  preds=pred.predictions.argmax(-1)# maxm value will be extracted
  acc=accuracy_score(labels,preds)
  return {
      'accuracy':acc
  }
```

```python
from transformers import Trainer
```

```python
trainer=Trainer(
    model=model,
    args=train_args,
    train_dataset=train_dataset_trainer,
    eval_dataset=test_dataset_trainer,
    compute_metrics=compute
)
```

```python
print("Starting the training process")
try:
  trainer.train()
  print('model has been trained successfully')
  trainer.save_model()
except Exception as e:
  print('model training was unsuccessful')
  print(e)
```

```python
SAVE_PATH="./finetuned_model"
```

```python
try:
  trainer.save_model(SAVE_PATH)
  tokenizer.save_pretrained(SAVE_PATH)
  print('model has been saved successfully')
except Exception as e:
  print('model saving was unsuccessful',e)
```

```python
finetuned_token=AutoTokenizer.from_pretrained(SAVE_PATH)
```

```python
finetuned_model=AutoModelForSequenceClassification.from_pretrained(SAVE_PATH)
```

```python
test_texts = [
    "This is a absolutely  wonderful",
    "I am  not happy with this situation",
    "Really poor quality service",
    "Absolutely fantastic product!",
]
```

```python
from transformers import pipeline
```

```python
pipe=pipeline('sentiment-analysis',model=finetuned_model,tokenizer=finetuned_token)
```
