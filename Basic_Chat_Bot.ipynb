```python
from transformers import AutoTokenizer,AutoModelForCausalLM # to dowanload tokenization pattern,model only...
```

```python
import torch
```

```python
model_name="deepseek-ai/deepseek-coder-1.3b-instruct"
```

```python
tokenizer=AutoTokenizer.from_pretrained(model_name)
```

```python
model_dep=AutoModelForCausalLM.from_pretrained(model_name)
```

```python
def generator(prompt,max_tokens=200):
  input=f"User: {prompt}\nAssistant" #Assistant should reply

  input=tokenizer(input,return_tensors='pt').to(model_dep.device)#to switch to GPU
  output=model_dep.generate(

              **input,
              max_new_token=max_tokens,
              temperature=0.7,
              do_sample=True,
              pad_token_id=tokenizer.eos_token_id
  )
  response= tokenizer.decode(output[0],skip_special_tokens=True)

  if 'Assistant' in response:
    response=response.split('Assistant')[-1].strip()
  else:
    response=''
  return response


```

```python
def generator(prompt,max_tokens=200):
  input=f"User: {prompt}\nAssistant" #Assistant should reply

  input=tokenizer(input,return_tensors='pt').to(model_dep.device)#to switch to GPU,but by default run on cpu..
  output=model_dep.generate(

              **input,# **select input of any size
              max_new_tokens=max_tokens,
              temperature=0.7,#restrict the creative power to controll hallucination 70%
              do_sample=True,#otherwise it will generate same answer every time for the question
              pad_token_id=tokenizer.eos_token_id#to stablize the model
  )
  response= tokenizer.decode(output[0],skip_special_tokens=True)#response is decoded

  if 'Assistant' in response:
    response=response.split('Assistant')[-1].strip()#extra space strip
  else:
    response=''
  return response

print("Welcome to the deepseek Chatbot Ask your Query")
print("Click quit or exit to end the chat")
while True:
  user=input("User :")

  if user.lower() in ['quit','exit']:
    print("Bye Nice Chating to you!!!")
    break

  response=generator(user)
  print("ChatBot:",response)
```
